```python
#Filtering dataset columns with only those columns which we would like to explore the data
df = girlsarst2020[
    [
        "Gender1",
        "AGE1",
        "EthnicGroup",
        "County",
        "Date",
        "Category1",
        "Offense1",
        "Region",
        "Stage",
        "Stages1",
    ]
]
df
```

```python
dfbk = df.copy()
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Offense_year"] = df["Date"].dt.year
df["Offense_month"] = df["Date"].dt.month
df["Offense_date"] = df["Date"].dt.day
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
ax = sns.countplot(
    df["EthnicGroup"],
    palette="winter_r",
    ax=ax,
    order=df["EthnicGroup"].value_counts().index,
)  # plotting first subplot
ax.set_title("Proportion of offenses by EthnicGroup")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
# plt.xlabel(f"Brand Name (Total unique brands {ub})")
plt.ylabel("Count")
# perc_annot(ax1,dfused['brand_name']) #perc_annot() function is used to show the percentages annotation
tot = len(df["EthnicGroup"])
for p in ax.patches:
    percentage = "{:.1f}%".format(
        100 * p.get_height() / tot
    )  # percentage of each class of the category
    x = p.get_x() + p.get_width() / 3  # width of the plot
    y = p.get_y() + p.get_height() + 1  # hieght of the plot
    ax.annotate(percentage, (x, y), size=20)  # annotate the percantage
```

### Observations:

- Black has the highest proportions of offenses

```python
fig, ax = plt.subplots(figsize=(15, 8))
ax = sns.countplot(df["Category1"], hue=df["EthnicGroup"], palette="winter_r")
tot = len(df["Category1"])
for p in ax.patches:
    percentage = "{:.1f}%".format(
        100 * p.get_height() / tot
    )  # percentage of each class of the category
    x = p.get_x() + p.get_width() / 5  # width of the plot
    y = p.get_y() + p.get_height() + 1  # hieght of the plot
    ax.annotate(percentage, (x, y), size=12)  # annotate the percantage
```

### Observations:

- Black has the highest proportions in all categories
- Misdemeanor is the highest offense category. ~50%

```python
fig, ax = plt.subplots(figsize=(20, 10))
# ax = sns.countplot(df["County"], hue=df["Category1"], palette="winter_r", orient="v")
ax = sns.barplot(df["County"], df["AGE1"])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax = sns.countplot(df["Offense1"], hue=df["EthnicGroup"])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

#### Observations:
